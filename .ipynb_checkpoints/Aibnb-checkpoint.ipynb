```python
data["available"].value_counts()
```

```python
def fct(group):
    return group.iloc[0]['date'] + "/" + group.iloc[-1]['date'] + "/" + str(group.shape[0]) + "/" + str(group.iloc[0]['listing_id']) + "/" + group.iloc[0]['available']
```

```python
adj_check = (data.available != data.available.shift()).cumsum()
newData = pd.DataFrame(data.groupby(['listing_id','available',adj_check], as_index=False, sort=False).apply(fct))
newData["start"],newData["end"],newData["num_day"],newData["listing_id"] ,newData["available"]  = newData[0].str.split('/',4).str
del newData[0]
newData
```
